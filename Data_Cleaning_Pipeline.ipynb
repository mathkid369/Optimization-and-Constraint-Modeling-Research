```python
import os
import json

# Path to your directory
PROBLEMS_DIR = "generated_updated_problems"

# Dictionary mapping Unicode math symbols ‚Üí English replacements
UNICODE_MAP = {
    "‚â§": " <= ",
    "‚â•": " >= ",
    "‚â†": " != ",
    "‚âà": " approximately ",
    "‚Üí": " to ",
    "‚Üê": " from ",
    "√ó": " times ",
    "¬∑": " * ",
    "‚àà": " in ",
    "‚àâ": " not in ",
    "‚äÜ": " subset of ",
    "‚äÇ": " proper subset of ",
    "‚äá": " superset of ",
    "‚äÉ": " proper superset of ",
    "‚à™": " union ",
    "‚à©": " intersection ",
    "‚àÄ": " for all ",
    "‚àÉ": " there exists ",
    "Œ£": " sum ",
    "Œ†": " product ",
    "Œª": " lambda ",
    "Œº": " mu ",
    "œÅ": " rho ",
    "Œ∏": " theta ",
    "Œ±": " alpha ",
    "Œ≤": " beta ",
    "Œ≥": " gamma ",
    "Œ¥": " delta ",
    "œÑ": " tau ",
    "œâ": " omega ",
    "‚àû": " infinity ",
    "¬±": " plus_or_minus ",
    "‚àÖ": " empty_set ",
    "‚Ä¶": " ... "
}

def clean_text(text):
    """Replace unicode chars using UNICODE_MAP"""
    for u, replacement in UNICODE_MAP.items():
        text = text.replace(u, replacement)
    return text

def clean_json_file(path):
    """Load, clean, and rewrite a JSON file"""
    try:
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)

        # Convert whole dict ‚Üí string ‚Üí clean ‚Üí back to dict
        as_text = json.dumps(data, ensure_ascii=False)
        cleaned = clean_text(as_text)
        cleaned_json = json.loads(cleaned)

        # Rewrite file
        with open(path, "w", encoding="utf-8") as f:
            json.dump(cleaned_json, f, indent=4, ensure_ascii=False)

        print(f"Cleaned {path}")

    except Exception as e:
        print(f"Error processing {path}: {e}")

def main():
    files = sorted(f for f in os.listdir(PROBLEMS_DIR) if f.endswith(".json"))

    for filename in files:
        full_path = os.path.join(PROBLEMS_DIR, filename)
        clean_json_file(full_path)

    print("Done cleaning all JSON files!")

if __name__ == "__main__":
    main()
```

```python
import os
import re
import json

PROBLEMS_DIR = "generated_updated_problems"

# Regex to detect unicode escape sequences like \u2208, \u03bb, etc.
UNICODE_ESCAPE_REGEX = re.compile(r"\\u[0-9a-fA-F]{4}")

def find_unicode_escapes_in_text(text):
    return UNICODE_ESCAPE_REGEX.findall(text)

def check_file_for_unicode(path):
    """Returns list of unicode escapes found in a JSON file."""
    with open(path, "r", encoding="utf-8") as f:
        raw = f.read()

    # Search the raw text directly (safe even if JSON has unicode escapes)
    return find_unicode_escapes_in_text(raw)

def main():
    print("üîç Scanning for Unicode escape sequences (\\uXXXX)...\n")

    for filename in sorted(os.listdir(PROBLEMS_DIR)):
        if not filename.endswith(".json"):
            continue

        full_path = os.path.join(PROBLEMS_DIR, filename)
        escapes = check_file_for_unicode(full_path)

        if escapes:
            print(f"‚ö†Ô∏è  {filename} contains {len(escapes)} unicode escapes:")
            print("    ", escapes)
        else:
            print(f"‚úÖ {filename} is clean (no \\uXXXX sequences).")

    print("\nDone scanning all files.")

if __name__ == "__main__":
    main()
```

```python
import os
import json
import py_compile
import matplotlib.pyplot as plt
import numpy as np

PROBLEMS_DIR = "generated_updated_problems"
SOLUTIONS_DIR = "generated_updated_solutions"

def try_compile(path):
    """Return True if Python file compiles, False otherwise."""
    try:
        py_compile.compile(path, doraise=True)
        return True
    except Exception:
        return False

def extract_model_type(problem_id):
    """Return ModelType from generated_updated_problems/problem_{id}.json."""
    json_path = os.path.join(PROBLEMS_DIR, f"problem_{problem_id}.json")
    if not os.path.exists(json_path):
        return None
    try:
        with open(json_path, "r", encoding="utf-8") as f:
            data = json.load(f)
        return data["Problem"]["ModelType"]
    except Exception:
        return None

def main():
    results = []             # True/False for each solution
    model_types = []         # ModelType for each solution
    modeltype_success = {}   # {ModelType: {"ok": x, "total": y}}

    for i in range(1, 501):
        sol_path = os.path.join(SOLUTIONS_DIR, f"solution_{i}.py")
        
        # Check compile success
        compiles = try_compile(sol_path)
        results.append(compiles)

        # Get corresponding ModelType
        model_type = extract_model_type(i)
        model_types.append(model_type)

        # Track stats
        if model_type not in modeltype_success:
            modeltype_success[model_type] = {"ok": 0, "total": 0}
        modeltype_success[model_type]["total"] += 1
        if compiles:
            modeltype_success[model_type]["ok"] += 1

    # Compute percentages
    summary = {}
    for mt, stats in modeltype_success.items():
        if stats["total"] > 0:
            pct = 100 * stats["ok"] / stats["total"]
        else:
            pct = 0.0
        summary[mt] = f"{pct:.1f}%"

    print("\n=== Compile Success by ModelType ===")
    print(json.dumps(summary, indent=2))

    # ---- GRID PLOT (20 √ó 30) ----
    grid = np.zeros((17, 30))  # 17 rows, 30 columns = 510 possible
    for idx, ok in enumerate(results):
        r = idx // 30
        c = idx % 30
        grid[r, c] = 1 if ok else 0   # green=1, red=0

    plt.figure(figsize=(12, 8))
    plt.title("Compilation Results for 500 Solutions")

    cmap = plt.cm.get_cmap("RdYlGn")  # red ‚Üí yellow ‚Üí green
    plt.imshow(grid, cmap=cmap, vmin=0, vmax=1)

    plt.colorbar(label="1 = compiles, 0 = fails")
    plt.xticks([])
    plt.yticks([])

    # Label each square with the solution number
    for idx in range(500):
        r = idx // 30
        c = idx % 30
        plt.text(c, r, str(idx + 1), ha="center", va="center", fontsize=5, color="black")

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
